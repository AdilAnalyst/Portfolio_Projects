# Visualization of Clean_data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('E:/Data analysis certificate stuff/Python/titanic_data.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['Survived'].value_counts()
```

```python
df['Sex'] = df['Sex'].apply(lambda x: 1 if x == 'male' else 0)
```

```python
df['Age'] = df['Age'].fillna(np.mean(df['Age']))
df.isnull().sum()
```

## Data Visualization

### 1. Counting the Survivors:

```python
plt.figure(figsize=(10,7))

#ax1 = sns.barplot(data=df, x='Survived', hue='Fare',palette='deep', legend=False )
plt.pie(df['Survived'].value_counts(), labels = ['Not Survived', 'Survived'], colors = ['#4594a4','#9ccbd4'], autopct="%1.1f%%")

plt.title('Total Survivors')

plt.show()
```

### 2. Survivors by Age

```python
plt.figure(figsize=(8, 6))

ax1 = sns.histplot(data=df, x='Age', hue='Survived',  multiple='stack', palette='deep', kde=False)

ax1.set_title('Survived People by Age')
ax1.set_xlabel('Age')
ax1.set_ylabel('Count')

plt.show()
```

The Highlighted one are the survivors and the rest (blue) ones are non-survivors

### 3. Count of gender

```python
plt.figure(figsize=(8, 6))

ax1 = sns.countplot(data=df, x='Sex', hue='Sex', palette='deep', legend=False)

ax1.set_title('People count by Gender')
ax1.set_xlabel('Gender')
ax1.set_ylabel('Count')

new_labels = ['Male', 'Female']
ax1.set_xticklabels(new_labels, rotation=45, fontsize = 10)

plt.show()
```

### 4. Survivors by gender

```python
plt.figure(figsize=(8, 6))

ax1 = sns.countplot(data=df, x='Sex', hue='Survived', palette='deep', legend=False)

ax1.set_title('Survivals count by Gender')
ax1.set_xlabel('Sex')
ax1.set_ylabel('Count')

new_labels = ['Male', 'Female']
ax1.set_xticklabels(new_labels, rotation=45, fontsize = 10)

plt.show()
```

### 5. Survivors by Class

```python
plt.figure(figsize=(8, 6))

ax1 = sns.countplot(data=df, x='Pclass', hue='Survived',palette='deep', legend=False)

ax1.set_title('Survivals count by Class')
ax1.set_xlabel('Class')
ax1.set_ylabel('Count')

plt.show()
```

### 6. Pair Plot

```python
plt.figure(figsize=(8, 6))

Features = ['Sex', 'Age', 'Fare', 'Pclass', 'Survived']
ax1 = sns.pairplot(df[Features], hue = 'Survived', palette= 'deep')

plt.suptitle('Pair Plot', y = 1.02)

plt.show()
```

### 7. Correlation Matrix 

```python
# Calculate the correlation matrix
corr_matrix = df[Features].corr()

# Set up the matplotlib figure
plt.figure(figsize=(24, 7))

# Create a new subplot for the correlation matrix
plt.subplots_adjust(right=0.9)  # Adjust space to make room for the heatmap
cax = plt.axes([1.05, 0.1, 0.2, 0.8])  # Define the position [left, bottom, width, height]

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=False, ax=cax)
cax.set_title('Correlation Matrix')

plt.show()
```
